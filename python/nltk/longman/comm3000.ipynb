```python
import nltk
import re
import pandas as pd
import numpy as np
import csv
from pprint import pprint
from IPython.display import display
```

## Read file

```python
filename = 'academic.txt'
lines = [x.strip() for x in open(filename).readlines() if len(x.strip()) > 0]
print(len(lines))
pprint(lines[-20:])
```

```python
data = [x.replace(',', '').split() for x in lines]
pprint(data[-20:])
```

```python
# list of number of items
lni = [len(x) for x in data]
print(lni[-10:])
print(max(lni))
print(min(lni))
```

```python
# list of number of items
lni = {}
for x in data:
    n = len(x)
    if n not in lni:
        lni[n] = 1
    else:
        lni[n] += 1
        
print(lni)
```

```python
target = [x for x in data if len(x) == 1]
pprint(target)
```

```python
# part of speech
pos = [
    'adj', # adjective
    'adv', # adverb
    'auxv', # auxiliary verb
    'conjunction', # conjunction
    'determiner', # determiner
    'interjection', # interjection
    'modv', # modal verb
    'n', # noun
    'number', # number
    'phrv', # phrasal verb
    'predeterminer', # predeterminer
    'prefix', # prefix
    'prep', # preposition
    'pron', # pronoun
    'suffix', # suffix
    'v', # verb
]
```

```python
res = []
for x in data:
    index = 2
    if x[1] in pos:
        index = 1
    
    word = ' '.join(x[:index])
    S, W = 0, 0
    for p in x[index + 1:]:
        if p[0] == 'S':
            S = int(p[1])
        if p[0] == 'W':
            W = int(p[1])
            
    res.append((word, x[index], S, W))

data = res
pprint(data[-20:])
```

```python
df = pd.DataFrame(data, index=np.arange(1, len(data) + 1), columns=['Word', 'POS', 'Spoken', 'Written'])
display(df)
```

```python
df.to_csv('academic.csv')
```

```python
from ipywidgets import *
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import display
%matplotlib inline
```

```python
x = np.linspace(0, 2 * np.pi)
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
line, = ax.plot(x, np.sin(x))
plt.close(fig)

def update(w = 1.0):
    line.set_ydata(np.sin(w * x))
    display(fig)
    
interact(update);
```

```python
x = np.linspace(0, 2 * np.pi)

def update(w = 1.0):
    fig = plt.figure()
    ax = fig.add_subplot(1, 1, 1)
    line, = ax.plot(x, np.sin(x))
    
    line.set_ydata(np.sin(w * x))
    fig.canvas.draw()
    
interact(update);
```

```python
from sympy import *
from sympy.plotting import plot, plot_parametric, plot_implicit
from IPython.display import display

init_printing(latex_mode='mathplotlib')
```

```python
x, y = symbols('x, y')
p = plot(sin(x), cos(x), show=False)
p.xlabel = 'time'
p.ylabel = ''
p.title = 'Sinosoide'
p.axis_center = (-1, -0.1)
p[0].line_color = [1, 0, 0]
p[1].line_color = [0, 1, 0]
p.show()
```

```python
plot_parametric(sin(x), cos(x), (x, 0, pi))
```

```python
plot_implicit(Eq(x**2 + y**2, 9))
```

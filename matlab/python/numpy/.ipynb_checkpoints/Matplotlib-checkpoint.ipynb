# Matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np
from pprint import pprint
%matplotlib inline
```

```python
x = [1, 3, 4, 5]
plt.plot(x)
plt.xlabel('x')
plt.ylabel('y')
plt.title('title')
plt.show()
```

```python
x = np.linspace(0, 2*np.pi, 1000)
y = np.sin(x);
plt.plot(x, y, 'r-.')
plt.axis([x.min(), x.max(), y.min(), y.max()])
plt.show()
```

```python
x = np.linspace(0, 2 * np.pi, 1000)
plt.plot(x, np.sin(x), 'r', x, np.cos(x), 'b')
plt.show()
```

```python
x = np.linspace(0, 2*np.pi)
plt.plot(x, np.sin(x), linewidth = 2.0);
```

```python
x = np.linspace(0, 2 * np.pi)
line = plt.plot(x, np.sin(x))
plt.setp(line, 'color', 'r', linewidth = 2.0)
plt.show()
plt.setp(line)
```

```python
x = np.linspace(0, 2 * np.pi)

plt.figure()

plt.subplot(211)
plt.plot(x, np.sin(x))

plt.subplot(212)
plt.plot(x, np.cos(x))
```

```python
x = np.linspace(0, 2 * np.pi, 10)
y = np.sin(x)
line = plt.plot(x, y)
plt.setp(line, 
        color=(1, 0, 0),
        linestyle='--',
        linewidth=2.0,
        marker='o',
        markersize=10.0,
        markeredgecolor=(0, 1, 0),
        markeredgewidth=2.0,
        markerfacecolor=(0, 0, 1))
plt.xlabel('x')
plt.ylabel('y')
plt.title('Sin')
plt.xlim((np.min(x) - 0.1, np.max(x) + 0.1))
plt.ylim((np.min(y) - 0.1, np.max(y) + 0.1))
print(line)
plt.show()
```

```python
x = np.linspace(0, 2 * np.pi, 10)
y = np.sin(x)

markerline, stemline, baseline = plt.stem(x, y)

plt.setp(markerline, color=(0, 1, 0), zorder=3.0)
plt.setp(stemline, linestyle='--')
plt.setp(baseline, linewidth=2.0)

plt.show()
```

```python
x = np.linspace(0, 2 * np.pi, 10)
y = np.sin(x)

line = plt.step(x, y)

plt.setp(markerline, color=(0, 1, 0), zorder=3.0)
plt.setp(stemline, linestyle='--')
plt.setp(baseline, linewidth=2.0)

plt.show()
```
